<a id=Problem Statement></a>
## Problem Statement

Given a dataset of features with education, income, and lifestyle related data find the best model and combination of features, with a twenty (20) feature maximum, to predict if a person's income is <=50K or >50K.

<a id=Executive Summary></a>
## Executive Summary
The team used a large dataset with 32,561 rows and 14 columns to train a prediction model. In total 6 classification models were run several times. In each round the number of features and hyperparameters were changed. Out of 3 rounds of 6 models, the best accuracy scores in the training and testing datasets are from random forest classifier model. Random forest classifier with the optimum hyperparameters returned 84% accuracy on the testing dataset. The team has to be very conscientious about the number of features and hyperparameters because of the constraint that was stated in the problem statement. The best performing model used 13 features.

```python
# Import libraries

import os
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", 100)

sns.set()
```

```python
# Read in dataset

dataset = pd.read_csv('./data/large_train_sample.csv')
```

```python
# View shape 

dataset.shape
```

```python
# Preview

dataset.head()
```

```python
# Check dtypes

dataset.dtypes
```

```python
# Veiw columns

dataset.columns
```

```python
# Check for nulls

dataset.isnull().sum()
```

```python
# Create list for categories

categorical_features = ['workclass', 'education',
       'marital-status', 'occupation', 'relationship', 'sex', 'native-country']
```

```python
# Create list for numeric types

numeric_features = [c for c in dataset.columns if c not in categorical_features]
numeric_features.remove('wage')
```

```python
# Remove spaces

dataset[categorical_features]=dataset[categorical_features].apply(lambda x: x.str.strip())
```

```python
# Replace ?s

for c in categorical_features:
    dataset[c]=dataset[c].replace('?','MISSING')
```

```python
# View unique values per feature

for c in categorical_features:
    print(f'{c} unique values: {dataset[c].unique()}')
```

```python
# Describe df

dataset.describe()
```

```python
# Count of categorical features

len(categorical_features)
```

```python
# Visualize unique values counts

fig, axes = plt.subplots(3, 3, figsize=(20, 18));
palette = iter(sns.husl_palette(len(categorical_features)))

for (feature, ax) in zip(categorical_features,axes.flatten()):
    plotter = sns.countplot(y=feature,data=dataset,ax=ax,color=next(palette),
                  orient='v',order=dataset[feature].value_counts().index);
    
    #plotter.set_xticklabels(plotter.get_xticklabels(),rotation=65)

    
plt.tight_layout()
```

```python
# Count of missing values

dataset[dataset['occupation']=='MISSING']['occupation'].value_counts()
```

```python
# Count of numeric features

len(numeric_features)
```

```python
# Visualize outliers

fig, axes = plt.subplots(2, 3, figsize=(20, 12));
palette = iter(sns.husl_palette(len(numeric_features)))

for (feature, ax) in zip(numeric_features,axes.flatten()):
    sns.boxplot(data=dataset, x=feature,ax=ax,color=next(palette));
    
ax.set_title('Boxplots of Numeric Features');    
```

```python
# Remove spaces

dataset['wage']=dataset['wage'].str.strip()
```

```python
# Convert to binary

dataset['wage']=dataset['wage'].map({'<=50K':0,'>50K':1})
```

```python
# Change dytpe

dataset['wage'].astype(int)
```

```python
# Create corr

corr_coef = dataset.corr()
```
